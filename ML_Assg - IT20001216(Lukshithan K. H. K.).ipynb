## Outlier Identification

```python
# This code creates a box plot using seaborn to display the distribution of values for each feature in the dataset. It uses 
# boxes to represent the IQR of the data, with whiskers indicating the range of the data, excluding outliers. The color scheme 
# is set by the palette parameter, and the x-axis labels are rotated using xticks. The plot is displayed with show()

plt.figure(figsize=(20,10))
sns.boxplot(data = data, palette = "Set2")
plt.xticks(rotation = 90)
plt.show()
```

The BoxPlots above show that the dataset contains many outliers, especially in the features histogram_variance, histogram_mode, histogram_median, and percentage_of_time_with_abnormal_long_term_variability. Consequently, we will start by removing the outliers.

## Outlier Removal

```python
# This code sets the lower and upper limits for the range of acceptable values for the "histogram_variance" feature based on 
# the mean and standard deviation of that feature. It then filters the data to only include values within this range, and 
# returns the number of rows in the filtered dataset. This is a method of removing outliers from the dataset

lowerlimit = data.histogram_variance.mean() - 3*data.histogram_variance.std()
upperlimit = data.histogram_variance.mean() + 3*data.histogram_variance.std()
data = data[(data.histogram_variance > lowerlimit) & (data.histogram_variance < upperlimit)]
data.shape[0]
```

```python
# This code removes outliers in the histogram_median attribute of the dataset. The lower and upper limits are defined using 
# the mean and standard deviation of the attribute, and any data points that fall outside of this range are removed from the 
# dataset. The resulting dataset shape is printed to show how many data points were removed

lowerlimit = data.histogram_median.mean() - 3*data.histogram_median.std()
upperlimit = data.histogram_median.mean() + 3*data.histogram_median.std()
data = data[(data.histogram_median > lowerlimit) & (data.histogram_median < upperlimit)]
data.shape[0]
```

```python
# This code calculates the lower and upper limit of acceptable values for the 
# percentage_of_time_with_abnormal_long_term_variability column in the dataset based on its mean and standard deviation. It 
# then filters out any rows where the percentage_of_time_with_abnormal_long_term_variability value falls outside this 
# acceptable range and updates the dataset with the filtered rows. Finally, it outputs the number of rows remaining in the 
# dataset after the filtering

lowerlimit = data.percentage_of_time_with_abnormal_long_term_variability.mean() - 3*data.percentage_of_time_with_abnormal_long_term_variability.std()
upperlimit = data.percentage_of_time_with_abnormal_long_term_variability.mean() + 3*data.percentage_of_time_with_abnormal_long_term_variability.std()
data = data[(data.percentage_of_time_with_abnormal_long_term_variability > lowerlimit) & (data.percentage_of_time_with_abnormal_long_term_variability < upperlimit)]
data.shape[0]
```

```python
# This code removes the outliers in the 'histogram_max' attribute of the 'data' dataset by defining a lower limit and an upper 
# limit as three standard deviations away from the mean. Then it filters the 'data' dataset to only include the rows where the 
# 'histogram_max' attribute is within the defined range. Finally, it prints the number of rows in the filtered 'data' dataset 
# using the shape[0] attribute.

lowerlimit = data.histogram_max.mean() - 3*data.histogram_max.std()
upperlimit = data.histogram_max.mean() + 3*data.histogram_max.std()
data = data[(data.histogram_max > lowerlimit) & (data.histogram_max < upperlimit)]
data.shape[0]
```

```python
# There are multiple lines of code where outliers are being removed for specific columns in the dataset by specifying a lower 
# and upper limit based on the mean and standard deviation of the column. The resulting dataset shape after removing the 
# outliers is also printed

lowerlimit = data.histogram_mode.mean() - 3*data.histogram_mode.std()
upperlimit = data.histogram_mode.mean() + 3*data.histogram_mode.std()
data = data[(data.histogram_mode > lowerlimit) & (data.histogram_mode < upperlimit)]
data.shape[0]
```

## Data Scaling and Splitting

```python
# This code imports the StandardScaler and train_test_split modules from the sklearn.preprocessing and sklearn.model_selection 
# libraries, respectively. These modules are commonly used for data preprocessing and model training in machine learning

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# the df_columns list is created to store the column names of the dataset. StandardScaler is then applied to the dataset to 
# scale the numerical features by removing the mean and scaling to unit variance. The scaled features are stored in a new 
# DataFrame called X

df_columns = ['baseline value', 'accelerations', 'fetal_movement',
       'uterine_contractions', 'light_decelerations', 'severe_decelerations',
       'prolongued_decelerations', 'abnormal_short_term_variability',
       'mean_value_of_short_term_variability',
       'percentage_of_time_with_abnormal_long_term_variability',
       'mean_value_of_long_term_variability', 'histogram_width',
       'histogram_min', 'histogram_max', 'histogram_number_of_peaks',
       'histogram_number_of_zeroes', 'histogram_mode', 'histogram_mean',
       'histogram_median', 'histogram_variance', 'histogram_tendency']
scale_X = StandardScaler()
X =  pd.DataFrame(scale_X.fit_transform(data.drop(["fetal_health"],axis = 1),), columns = df_columns)
X.head()
```

```python
y = data["fetal_health"] # extracts the column "fetal_health" from the "data" DataFrame and assigns it to the variable "y"
```

```python
# This code splits the dataset into training and testing sets using the train_test_split function from the 
# sklearn.model_selection module. The X variable represents the standardized input features, and the y variable represents the 
# target variable, "fetal_health". The function splits the data into 70% training and 30% testing sets, with a random state of '
# 42 for reproducibility. The stratify parameter is used to ensure that the target variable is equally represented in both the 
# training and testing sets. The resulting variables are X_train, X_test, y_train, and y_test, which represent the input and 
# target variables for the training and testing sets, respectively

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42, stratify = y)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Random Forest Regression Application

```python
# imports the RandomForestClassifier class from the sklearn.ensemble module. The RandomForestClassifier is a machine learning 
# model that uses an ensemble of decision trees to make predictions

from sklearn.ensemble import RandomForestClassifier
```

```python
# fits a Random Forest Classifier model on the training data (X_train and y_train), and then evaluates the accuracy of the model
# on the test data (X_test and y_test). The accuracy is printed to the console using f-strings. Finally, the model is used to 
# make predictions on the test data, and the predicted values are stored in the variable pred_random_forest

random_forest = RandomForestClassifier()
random_forest_mod = random_forest.fit(X_train, y_train)
print(f"Baseline Random Forest: {round(random_forest_mod.score(X_test, y_test), 3)}")

pred_random_forest = random_forest_mod.predict(X_test)
```

```python
# two functions from the sklearn.metrics module that are commonly used to evaluate the performance of a machine learning model

from sklearn.metrics import classification_report, confusion_matrix
```

```python
# This code is using the classification_report function from the sklearn.metrics library to print the classification report for 
# the predictions made by the Random Forest model on the test set. The classification report shows several metrics for each 
# class, such as precision, recall, and F1-score, as well as the overall accuracy of the model

print("Classification Report")
print(classification_report(y_test, pred_random_forest))
```

```python
# This code creates a heatmap of the confusion matrix, showing the number of correct and incorrect predictions made by the 
# random forest model on the test set. The x-axis represents the predicted values, while the y-axis represents the actual 
# values. The numbers in each cell indicate the count of observations that fall into that category

ax= plt.subplot()
sns.heatmap(confusion_matrix(y_test, pred_random_forest), annot=True, ax = ax, cmap = "BuPu");

ax.set_xlabel("Predicted labels"); # x-axis label
ax.set_ylabel("True labels");      # y-axis label
ax.set_title("Confusion Matrix");  # title
ax.xaxis.set_ticklabels(["Normal", "Suspect", "Pathological"]); # ticks
```
